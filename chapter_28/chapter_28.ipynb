# 层次聚类应用及聚类树绘制


```python
# 加载小麦数据集
import pandas as pd

df = pd.read_csv("challenge-8-seeds.csv")
df.head()
```

```python
# Exercise 28.1
# 使用 SciPy 中的 Agglomerative 聚类方法完成小麦种子层次聚类
from scipy.cluster import hierarchy

### 代码开始 ### (≈ 1 行代码)
Z = hierarchy.linkage(df, method ='ward', metric='euclidean')
### 代码结束 ###
Z[-1]
```

可以看到scipy把整个层次聚类的过程体现出来了

```python
# Exercise 28.2 
# 使用 SciPy 中的 dendrogram 方法绘制小麦种子层次聚类二叉树
from matplotlib import pyplot as plt
%matplotlib inline

plt.figure(figsize=(15, 8))
### 代码开始 ### (≈ 1 行代码)
hierarchy.dendrogram(Z)
### 代码结束 ###
plt.show()
```

```python
# Exercise 28.3
# 对小麦种子层次聚类二叉树进行修剪
plt.figure(figsize=(15, 8))
### 代码开始 ### (≈ 1 行代码)
hierarchy.dendrogram(Z, truncate_mode='lastp', p=15, show_leaf_counts=True, show_contracted=True)
### 代码结束 ###
plt.show()
```
